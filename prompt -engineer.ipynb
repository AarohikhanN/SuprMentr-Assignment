```python
prompts = {
    "Resume": {
        "Weak": "Write a resume",
        "Strong": "Write a professional resume for a fresher data analyst with skills in Python, SQL, and machine learning"
    },
    "Business Idea": {
        "Weak": "Give me a business idea",
        "Strong": "Suggest a low-budget online business idea for students in India with growth potential"
    },
    "Study Plan": {
        "Weak": "Make a study plan",
        "Strong": "Create a 30-day study plan for learning Python for beginners with daily tasks"
    }
}

for category, prompt in prompts.items():
    print(f"\n{category}:")
    print("Weak Prompt:", prompt["Weak"])
    print("Strong Prompt:", prompt["Strong"])
```
