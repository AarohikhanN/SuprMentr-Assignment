```python
import string

# Sample stopwords list
stopwords = ["is", "am", "the", "this", "for", "i", "u", "me"]

def clean_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))

    # Split words
    words = text.split()

    # Remove stopwords
    words = [word for word in words if word not in stopwords]

    # Join back
    cleaned_text = " ".join(words)

    return cleaned_text

# Test
sentence = "This is soooo amazing!!! 😍"
print("Original:", sentence)
print("Cleaned:", clean_text(sentence))
```
