```python
import re

# Sample resume text
resume = """
I am a Python developer with knowledge in machine learning, data analysis,
and SQL. I have worked on NLP projects and web development.
"""

# Job required skills
job_skills = ["python", "machine learning", "data analysis", "sql", "nlp"]

# Preprocessing function
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text

# Analyze resume
def analyze_resume(resume_text, skills):
    resume_text = clean_text(resume_text)
    found_skills = []

    for skill in skills:
        if skill in resume_text:
            found_skills.append(skill)

    match_score = (len(found_skills) / len(skills)) * 100

    return found_skills, match_score

# Run analysis
skills_found, score = analyze_resume(resume, job_skills)

# Output
print("Skills Found:", skills_found)
print("Match Score:", score, "%")

if score > 70:
    print("Good match for the job!")
elif score > 40:
    print("Average match, improve skills.")
else:
    print("Low match, needs improvement.")
```
