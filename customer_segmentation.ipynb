```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Example user ratings
data = np.array([
    [5, 1, 2],  # User A
    [4, 1, 2],  # User B
    [1, 5, 4]   # User C
])

model = NearestNeighbors(n_neighbors=2)
model.fit(data)

distances, indices = model.kneighbors([data[0]])

print("Nearest neighbors of User A:", indices)
```
