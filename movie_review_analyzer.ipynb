```python
from textblob import TextBlob

reviews = [
    "The movie was fantastic and amazing",
    "I hated the film, it was boring",
    "It was an average movie",
    "Absolutely loved the acting",
    "Worst movie ever"
]

for review in reviews:
    blob = TextBlob(review)
    sentiment = blob.sentiment.polarity

    if sentiment > 0:
        result = "Positive"
    elif sentiment < 0:
        result = "Negative"
    else:
        result = "Neutral"

    print(f"Review: {review}")
    print(f"Sentiment: {result}\n")
```
