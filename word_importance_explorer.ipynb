```python
from sklearn.feature_extraction.text import TfidfVectorizer

# 5 sample documents
docs = [
    "Machine learning is amazing",
    "Python is used for machine learning",
    "AI and machine learning are the future",
    "Deep learning is a part of AI",
    "Python makes AI easy"
]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(docs)

feature_names = vectorizer.get_feature_names_out()

# Print top words for each document
for i, doc in enumerate(tfidf_matrix.toarray()):
    print(f"\nDocument {i+1} Top Words:")
    top_indices = doc.argsort()[-3:][::-1]
    for index in top_indices:
        print(feature_names[index])
```
